## 1. Loading the packages and Importing the data.

Now I will import the Pandas, NumPy and MatPlotLib libraries and load the trees.csv dataset into the variable dt.

```python
# Loading packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Importing our dataset with new column names
dt = pd.read_csv('/Users/diogo/Desktop/trees.csv', names = ['Index','Girth','Height','Volume'])
dt
```

As we can see, the dataset has 4 columns and 31 rows. <br> Although it needs cleaning, and that's what will be done in the following step.

## 2. Data Cleaning

First of all, we loaded the dataset with the already corrected names for our collumns. However, that means that the 1st row of our dataset are the previous names for our columns. 
We notice as well that we have 2 indexes with the same exact numbers, we can delete our index column.

```python
# Deleting the first row of the dataset
dt.drop(0, inplace=True)

# Deleting the Index column
del dt['Index']

dt
```

As expected, the 0 row and the Index column have been deleted and we have a better dataset. 

Now we need to take care of our data. If you take a look at the row we just eliminated (first code block), *Girth* is in **inches**, *Height* is in **feet** and *Volume* is in **cubic feet**. To set them accordingly, we are going to transform them all to meters.

1 inch = 0.0254 meters <br>
1 foot = 0.3048 meters <br>
1 cubic foot = 0.0283 meters <br><br>Afterwards I will round them all to 1 decimal point to keep the data clean.

```python
# Transform all the data into float type in order to multiply it.
dt['Girth'] = dt['Girth'].astype('float')
dt['Height'] = dt['Height'].astype('float')
dt['Volume'] = dt['Volume'].astype('float')

# Next I will multiply them by the transforming values.

dt['Girth'] = dt['Girth'] * 0.0254
dt['Height'] = dt['Height'] *0.3048
dt['Volume'] = dt['Volume'] *0.0283

# Round all of them to 1 decimal point.

dt['Girth'] = dt['Girth'].round(1)
dt['Height'] = dt['Height'].round(1)
dt['Volume'] = dt['Volume'].round(1)

dt
```

This concludes our data cleaning process. Following we will measure our data.

## 3. Data Measuring

First of all let us find out all the maximum and minimum values.

```python
# Finding the maximum values
max_girth = dt['Girth'].max()
max_height = dt['Height'].max()
max_volume = dt['Volume'].max()

# Finding the minimum values
min_girth = dt['Girth'].min()
min_height = dt['Height'].min()
min_volume = dt['Volume'].min()

print('The maximum Girth is '+ str(max_girth)+ ' meters and the minimum Girth is ' + str(min_girth)+' meters')
print('The maximum Height is '+str(max_height)+' meters and the minimum Height is '+str(min_height)+' meters')
print('The maximum Volume is '+str(max_volume)+' cubic meters and the minimum Volume is '+str(min_volume)+' cubic meters')
```

Now that we know our maximums and minimums, it's time to go into the statistical properties of the dataset.<br>Mean, Mode and Median.

```python
# For Girth
avg_girth = round(dt['Girth'].mean(),2)
med_girth = round(dt['Girth'].median(),2)
mod_girth = round(dt['Girth'].mode(),2)
# For Height
avg_height = round(dt['Height'].mean(),2)
med_height = round(dt['Height'].median(),2)
mod_height = round(dt['Height'].mode(),2)
# For Volume
avg_volume = round(dt['Volume'].mean(),2)
med_volume = round(dt['Volume'].median(),2)
mod_volume = round(dt['Volume'].mode(),2)

print('The mean for Girth is '+str(avg_girth)+', the median for Girth is '+str(med_girth)+' and the mode for Girth is '+str(mod_girth))

print('The mean for Height is '+str(avg_height)+', the median for Height is '+str(med_height)+' and the mode for Height is '+str(mod_height))

print('The mean for Volume is '+str(avg_volume)+', the median for Volume is '+str(med_volume)+' and the mode for Volume is '+str(mod_volume))
```

Now we know the statistical values, we will proceed to visualize the data.

## 4. Data Visualization

Now we will proceed with the creation of Graphs.

First, one for Girth

```python
plt.plot(dt['Girth'])
plt.title('Girth')
plt.ylabel('Meters')
plt.xlabel('Index')
plt.show()
```

Now for Height

```python
plt.plot(dt['Height'])
plt.title('Height')
plt.ylabel('Meters')
plt.xlabel('Index')
plt.show()
```

And finaly, Volume

```python
plt.plot(dt['Volume'])
plt.title('Volume')
plt.ylabel('Cubic Meter')
plt.xlabel('Index')
plt.show()
```

Now we are ready to create our final report!

## 5. Final Report

In this final report, I will create a dataframe from scratch with the statistical data and provide the graphs.

```python

# Creating a table with the statistical data
df = pd.DataFrame({'Girth' : [max_girth, min_girth, avg_girth, med_girth, mod_girth[0]], 'Height' : [max_height, min_height, avg_height, med_height, mod_height[0]], 'Volume' : [max_volume, min_volume, avg_volume, med_volume, mod_volume[0]]}, index = ['Maximum', 'Minimum', 'Mean', 'Median', 'Mode'])

# Displaying the original data and the new table
print('Tree measurements data\n')
display(dt)

print ('Measurements\n')
display(df)

# Recreating the graphs
# Girth
plt.plot(dt['Girth'])
plt.title('Girth')
plt.ylabel('Meters')
plt.xlabel('Index')
plt.show()
# Height
plt.plot(dt['Height'])
plt.title('Height')
plt.ylabel('Meters')
plt.xlabel('Index')
plt.show()
# Volume
plt.plot(dt['Volume'])
plt.title('Volume')
plt.ylabel('Cubic Meter')
plt.xlabel('Index')
plt.show()
```
